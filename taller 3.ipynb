Se necesita un sistema que pida un sueldo de un
trabajador, si el sueldo es mayor a 655000, no tedrá
ninguna bonificación, pero si es menor tiene bonificación
del 4%. Al final se desea saber el total del sueldo con
o sin bonificación. Preguntar si desea ejecutar de nuevo
el programa con el ciclo while.

```python
# Definimos una variable para ejecutar el ciclo while
ejecutar_programa = True

# Iniciamos el ciclo while
while ejecutar_programa:

    
    sueldo = float(input("Ingrese el sueldo del trabajador: "))

    
    if sueldo > 655000:
        print("Este sueldo no tiene bonificación.")
        total_sueldo = sueldo
    else:
       
        bonificacion = sueldo * 0.04
        total_sueldo = sueldo + bonificacion
        print(f"El sueldo con bonificación es: {total_sueldo:,.0f}")

    print(f"El total del sueldo es: {total_sueldo:,.0f}")

   
    respuesta = input("¿Desea ejecutar el programa de nuevo? (s/n): ")

   
    if respuesta.lower() == "s":
        ejecutar_programa = True
    else:
        ejecutar_programa = False


print("¡Gracias por utilizar nuestro programa!")
```

Crear un bucle que cuente todos los números pares hasta
el 100 ciclo for

```python
# Utilizamos un ciclo for para iterar sobre los números del 0 al 100
for num in range(0, 101):
    # Verificamos si el número es par
    if num % 2 == 0:
        # Imprimimos el número si es par
        print(num)
```
